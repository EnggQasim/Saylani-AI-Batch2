```python
score = [10,20,50,23,0,50,20,60,10]
fig, ax = plt.subplots()
ax.hist(score, bins=25, edgecolor='black')

labels = [item.get_text() for item in ax.get_xticklabels()] 

def update_ticks(label):
    l1 = []
    for i in range(len(labels)):
        if i == 1:
            l1.append('low')
        elif i == len(labels)//2:
            l1.append('Average')
        elif i == 8:
            l1.append('high')
        else:
            l1.append(i)
    return l1

ax.set_xticklabels(update_ticks(labels))

plt.show()
```

```python
import matplotlib.ticker as mticker
import numpy as np 

def update_ticks(x, pos):
    if x == 0:
        return 'Mean'
    else:
        return x

data = np.random.normal(0, 1, 1000)
fig, ax = plt.subplots()
ax.hist(data, bins=25, edgecolor='black')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(update_ticks))
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x=[2,4,3,4,6,5,3,5,6]
y=[1,3,5,6,6,7,7,8,9]
labels = [x+2 for x in y]

plt.scatter(x,y, label="data1", color='r', s=100, marker="*")
for label, x_count, y_count in zip(labels, x, y):
    plt.annotate(label,
                 xy=(x_count, y_count), # put the label with its point
                 xytext=(5, 5), # but slightly offset
                 textcoords='offset points')

plt.title("PIAIC Demo")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```
