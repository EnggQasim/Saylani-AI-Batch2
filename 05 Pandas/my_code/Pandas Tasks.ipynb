### Task 2
create a DataFrame of no: of column equal to the excel number of column and then change their column names according to the excel sheet columns name

```python
arr = np.ones((10,16384))
df2 = pd.DataFrame(arr)
df2.head()
```

```python
def excel_col(col):
    """Covert 1-relative column number to excel-style column label."""
    quot, rem = divmod(col-1,26)
    return excel_col(quot) + chr(rem+ord('A')) if col!=0 else ''
```

```python
lst = []
for i in [i for i in range(16384)]:
    lst.append(excel_col(i+1))
```

```python
lst
```

```python
df2 = pd.DataFrame(arr, columns= lst)
df2
```
