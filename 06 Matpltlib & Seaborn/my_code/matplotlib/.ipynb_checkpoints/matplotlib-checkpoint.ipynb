```python
x_values = [-0.5, 0.0, 0.5]
y_values = [-0.5, 0.0, 0.5]

# subplot 1
plt.subplot(2,3,1)
plt.plot(x_values, y_values, ls= '', marker= 's')
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 1')

#subplot 2
plt.subplot(2,3,2)
plt.plot(x_values, y_values, ls= '', marker= '*')
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 2')

#subplot 3
plt.subplot(2,3,3)
plt.plot(x_values, y_values, ls= '', marker= 'o')
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 3')

#subplot 4
plt.subplot(2,3,4)
plt.plot(x_values, y_values, ls= '', marker= 'o', ms= 13)
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 4')

#subplot 5
plt.subplot(2,3,5)
plt.plot(x_values, y_values, ls= '', marker= 's', ms= 10)
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 5')

#subplot 6
plt.subplot(2,3,6)
plt.plot(x_values, y_values, ls= '', marker= 'p', ms= 15)
plt.axis([-0.4, 0.4, -0.4,0.4 ])
plt.title('subplot 6')

# subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
# left  = 0.125  # the left side of the subplots of the figure
# right = 0.9    # the right side of the subplots of the figure
# bottom = 0.1   # the bottom of the subplots of the figure
# top = 0.9      # the top of the subplots of the figure
# wspace = 0.2   # the amount of width reserved for blank space between subplots
# hspace = 0.2   # the amount of height reserved for white space between subplots

plt.subplots_adjust( wspace=0.7, hspace=0.7)
plt.suptitle('A Figure with 6 subplots', horizontalalignment= 'center', fontsize= 15)
```

```python
# establish data
x = [1, 2, 3]
y = [4, 5, 6]

figure, axes = plt.subplots(nrows=2, ncols=2)

# create subplots
axes[0, 0].plot(x, y, marker= 's')
axes[0, 1].plot(x, y, ls= '--')
axes[1, 0].plot(x, y, ls= ':')
axes[1, 1].plot(x, y, marker= '*', ms= 10)

figure.suptitle("Main Title")
```

```python
fig, axs = plt.subplots(1, 2, constrained_layout=True)

ax = axs[0] #<-- 1-dimentional  signal argument
ax.plot(range(10))
# ax.xaxis.set_label_position('top') #<-- setting x_label position onn top
ax.set_xlabel('X-label')
ax.set_title('Center Title')

ax = axs[1] # <-- 1-dimentioanl 
ax.plot(range(10))
ax.xaxis.set_label_position('top') 
ax.xaxis.tick_top()   #<-- setting x_label position onn top
ax.set_xlabel('X-label')
ax.set_title('My Title', y=1.0, pad=-14)
plt.show()
```

```python
score = [10,20,50,23,0,50,20,60,10]
fig, ax = plt.subplots()
ax.hist(score, bins=25, edgecolor='black')

labels = [item.get_text() for item in ax.get_xticklabels()] 

def update_ticks(label):
    l1 = []
    for i in range(len(labels)):
        if i == 1:
            l1.append('low')
        elif i == len(labels)//2:
            l1.append('Average')
        elif i == 8:
            l1.append('high')
        else:
            l1.append(i)
    return l1

ax.set_xticklabels(update_ticks(labels))

plt.show()
```

```python
import matplotlib.ticker as mticker
import numpy as np 

def update_ticks(x, pos):
    if x == 0:
        return 'Mean'
    else:
        return x

data = np.random.normal(0, 1, 1000)
fig, ax = plt.subplots()
ax.hist(data, bins=25, edgecolor='black')
ax.xaxis.set_major_formatter(mticker.FuncFormatter(update_ticks))
plt.show()
```

### Annotation

```python
x=[2,4,3,4,6,5,3,5,6]
y=[1,3,5,6,6,7,7,8,9]
labels = [x+2 for x in y]

plt.scatter(x,y, label="data1", color='r', s=100, marker="*")
for label, x_count, y_count in zip(labels, x, y):
    plt.annotate(label,
                 xy=(x_count, y_count), # put the label with its point
                 xytext=(5, 5), # but slightly offset
                 textcoords='offset points')

plt.title("PIAIC Demo")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```
