```python
with open('abc3.txt', 'r+') as f:
    a = f.readlines()
    a.reverse()
    
    b = ['A\n','B\n','C\n']
    b.reverse()
    
    c = b + a
    d = "".join(c)
    print(d)
    f.seek(0)
    f.write(d)
    
    f.seek(0)
    print(f.readlines())
```

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)
 
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
 
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
